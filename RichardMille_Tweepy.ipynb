```python
def text_clean(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub(r'http\S+', '', text, flags = re.MULTILINE)
    return text

cleaned = lambda x: text_clean(x)

cleaned_text = lambda x: text_clean(x)
```
